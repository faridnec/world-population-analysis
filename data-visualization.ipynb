# Data Visualization

Various visualizations are done in order to learn pattern within the data

```python

import matplotlib.pyplot as plt ## Import essential libraries
import pandas as pd
import seaborn as sns
```

```python
 ## Read csv file and save into dataframe
df = pd.read_csv("./data/world_population_data.csv")
```

```python
## Create a dictionary to rename the columns
new_cols_dict = { 
    'cca3' : 'country_code',
    'area (km²)' : 'area',
    'density (km²)' : 'density',
    'growth rate' : 'growth_rate',
    'world percentage' : 'world_percentage'
}
df.rename(new_cols_dict, axis = 1, inplace= True) ## Rename the columns
```

```python
## Export the file as excel
df.to_excel('world_population_data.xlsx'    , sheet_name= 'Data') 
```

Filter the dataframe to obtain countries in Europe and population more than 10 million

```python
## European countries with population more than 10 million
country_x = df[(df['2023 population'] > 10000000) & (df['continent'] == 'Europe')] 
country_x
```

Create a scatter plot for the countries we filtered above
We used the populations for 1970, 2000, and 2023

```python
## Create a scatter plot the for the countries we filtered above 
## The plot shows the populations in 1970, 2000, and 2023
plt.figure(figsize= (10, 6)) 
plt.scatter(country_x['country'], country_x['1970 population'], color = 'green', s = 70) 
plt.scatter(country_x['country'], country_x['2000 population'], color = 'blue', s = 70)
plt.scatter(country_x['country'], country_x['2023 population'], color = 'red', s = 70)
plt.ylabel('Population (million)', fontweight = 'bold', size = 9, labelpad= 15)
plt.title('Population (1970 - 2000 - 2023)', fontweight = 'bold', pad = 10)
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.3)
plt.xticks(rotation=45, ha='right',  size = 12)
labels = ('1970', '2000', '2023')
plt.legend(labels = labels, fontsize = 'large', loc = 'upper right', bbox_to_anchor = (0.98, 0.98))
plt.tight_layout()
plt.show()
```

Pie chart represents the world percentage (population) of the countries

```python
## Create a copy of the dataframe and remove the percentage from the world_percentage column.
## Create a pie chart based on continents
df_a = df.copy()
df_a['world_percentage'] = df_a['world_percentage'].str.rstrip('%')
df_a['world_percentage'] = df_a['world_percentage'].astype(float)
continent_totals = df_a.groupby('continent')['world_percentage'].sum().reset_index()
plt.figure(figsize=(8, 8))
cols = ['#E8D63E', '#1B8E43', '#F15331','#D1DEB8', 'blue', 'm']
plt.pie(continent_totals['world_percentage'], labels=continent_totals['continent'], autopct='%1.1f%%', startangle=30, colors = cols)
plt.title('Total World Percentage Distribution by Continent', fontdict= {'fontsize': 20, 'fontweight' : 'bold'}, y = 0.93, pad = 5)
plt.legend(loc='lower left', bbox_to_anchor=(0.2, 0.5), title = 'Continents', title_fontsize = 12)
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.size'] = 13
plt.tight_layout()
plt.show()
```

Column named population difference added to the dataframe

```python
## Create a new column called population difference by subtracting 1970's population from 2023's population
df_a['population_difference'] = df_a['2023 population'] - df_a['1970 population']
df_a
```

Line plots for the countries with initial A,B, and C
Visualized by country and their density

```python
## Find the countries start with letter A, B, and C
a_country = df_a[df_a['country'].str.startswith('A')][['country', 'population_difference', 'growth_rate', 'density', 'area']].reset_index()
a_country['growth_rate'] = a_country['growth_rate'].str.rstrip('%')
b_country = df_a[df_a['country'].str.startswith('B')][['country', 'population_difference', 'growth_rate', 'density', 'area']].reset_index()
b_country['growth_rate'] = a_country['growth_rate'].str.rstrip('%')
c_country = df_a[df_a['country'].str.startswith('C')][['country', 'population_difference', 'growth_rate', 'density', 'area']].reset_index()
c_country['growth_rate'] = a_country['growth_rate'].str.rstrip('%')

## Create subplots by the country and their density
plt.figure(figsize = (25, 8))
plt.subplot(131)
plt.plot(a_country['country'], a_country['density'], color = 'r', linewidth = 3)
plt.xticks(rotation = 45, ha = 'right', size = 18)
plt.ylabel('(Density)', fontweight = 'bold', size = 9, labelpad= 10, loc = 'center')
plt.grid(True, linewidth = 0.3, linestyle = 'dashed')
plt.title('Countries with initial A', fontdict= {'fontsize' : 20, 'fontweight' : 'bold'}, pad = 15 )

plt.subplot(132)
plt.plot(b_country['country'], b_country['density'], color = 'b', linewidth = 3)
plt.xticks(rotation = 45, ha = 'right', size = 18)
plt.ylabel('(Density)', fontweight = 'bold', size = 9, labelpad= 10, loc = 'center')
plt.grid(True, linewidth = 0.3, linestyle = 'dashed')
plt.title('Countries with initial B', fontdict= {'fontsize' : 20, 'fontweight' : 'bold'}, pad = 15 )

plt.subplot(133)
plt.plot(c_country['country'], c_country['density'], color = 'g', linewidth = 3)
plt.xticks(rotation = 45, ha = 'right', size = 18)
plt.ylabel('(Density)', fontweight = 'bold', size = 9, labelpad= 10, loc = 'center')
plt.title('Countries with initial C', fontdict= {'fontsize' : 20, 'fontweight' : 'bold'}, pad = 15 )
plt.grid(True, linewidth = 0.3, linestyle = 'dashed')

# Display the plt
plt.suptitle('Density by countries', family = 'Times New Roman', fontweight = 'bold',  fontsize = 40,  y= 1)
plt.tight_layout()
plt.show()
```

### Visualizing Distribution of Population Based on Continents

```python
# Plot distribution of population based on continents
from matplotlib.ticker import FuncFormatter


plt.figure(figsize=(10, 6))
sns.barplot(x='continent', y='2023 population', data=df, estimator=sum, errorbar=None)
plt.title('Distribution of Population Based on Continents (2023)')
plt.ylabel('Total Population')

# Format y-axis ticks as billions with a thousand separator
def billions_formatter(x, pos):
    return f'{x / 1e9:.1f}B'

plt.gca().yaxis.set_major_formatter(FuncFormatter(billions_formatter))
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```
