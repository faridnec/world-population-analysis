```python
new_cols_dict = {
    'cca3' : 'country_code',
    'area (km²)' : 'area',
    'density (km²)' : 'density',
    'growth rate' : 'growth_rate',
    'world percentage' : 'world_percentage'
}
df.rename(new_cols_dict, axis = 1, inplace= True)
```

```python
country_x = df[(df['2023 population'] > 10000000) & (df['continent'] == 'Europe')]
country_x
```

```python
plt.figure(figsize= (10, 6))
plt.scatter(country_x['country'], country_x['1970 population'], color = 'green', s = 70) 
plt.scatter(country_x['country'], country_x['2000 population'], color = 'blue', s = 70)
plt.scatter(country_x['country'], country_x['2023 population'], color = 'red', s = 70) 
plt.ylabel('Population (million)', fontweight = 'bold')
plt.title('Population (1970 - 2000 - 2023)', fontweight = 'bold')
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.3)
plt.xticks(rotation=45, ha='right')
labels = ('1970', '2000', '2023')
plt.legend(labels = labels, fontsize = 'large', loc = 'upper right', bbox_to_anchor = (0.98, 0.98))
plt.tight_layout()
plt.show()
```

```python
continent_totals = df.groupby('continent')['world_percentage'].sum().reset_index()
plt.figure(figsize=(8, 8))
cols = ['#E8D63E', '#1B8E43', '#F15331','#D1DEB8', 'blue', 'm']
plt.pie(continent_totals['world_percentage'], labels=continent_totals['continent'], autopct='%1.1f%%', startangle=30, colors = cols)
plt.title('Total World Percentage Distribution by Continent', fontdict= {'fontsize': 20, 'fontweight' : 'bold'}, y = 0.93)
plt.legend(loc='lower left', bbox_to_anchor=(0.2, 0.5), title = 'Continents', title_fontsize = 12)
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.size'] = 10 
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Example: df = pd.read_csv('your_data.csv')

# Select relevant columns for the heatmap
heatmap_data = df[['country', 'continent', 'world_percentage', 'growth_rate']]
# Pivot the data for the heatmap
heatmap_data_pivot = heatmap_data.pivot(index='continent', columns='growth_rate', values='world_percentage')

# Create a heatmap using seaborn
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(heatmap_data_pivot, annot=True, cmap='viridis', fmt=',.0f', linewidths=.5)
plt.title('Heatmap for 2023 Population by Country and Continent', fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.show()
```

```python
hp = df[['world_percentage','density','area', '2023 population', 'country', 'continent']]
hp_pivot = hp.pivot(index = 'continent', columns = 'area', values = 'density')
plt.figure(figsize=(20, 20))
heatmap = sns.heatmap(hp_pivot, annot = True, cmap='viridis', fmt=',.0f', linewidths=.5)
plt.title('Heatmap for 2023 Population by Country and Continent', fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.show()
```
